```python
import time

from src.environments import StochasticGridEnvironment
from src.agents import DQNAgent
from src.value_functions import ValueFunction
from src.utils import run_env
import matplotlib.pyplot as plt
import numpy as np
import pickle
```

```python
env = StochasticGridEnvironment(max_timesteps=30)
```

#### DQN Analysis

```python
with open('sadapala_assignment2_part2_dqn_gridworld.pickle', 'rb') as f:
    dqn_agent = pickle.load(f)
```

```python
epsilons = []
epsilon = 1
epsilon_decay_factor = np.power(0.01/epsilon, 1/1200)
for _ in range(800):
    epsilons.append(epsilon)
    epsilon = epsilon * epsilon_decay_factor
```

```python
plt.plot(epsilons)
plt.xlabel('Episode')
plt.ylabel('Epsilon')
plt.title('Epsilon Across Episodes')
plt.show()
```
